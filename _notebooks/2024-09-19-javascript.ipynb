  <button id="changeColorBtn">Change Background Color</button>
  <button id="counterBtn">Increase Counter</button>
  <button id="greetBtn">Greet Me</button>

  <div id="counter">Counter: 0</div>
  <div id="message"></div>

  <script>
    // Change background color on button click
    const changeColorBtn = document.getElementById('changeColorBtn');
    changeColorBtn.addEventListener('click', function() {
      const randomColor = '#' + Math.floor(Math.random()*16777215).toString(16);
      document.body.style.backgroundColor = randomColor;
    });

    // Counter button functionality
    const counterBtn = document.getElementById('counterBtn');
    let counterValue = 0;
    counterBtn.addEventListener('click', function() {
      counterValue++;
      document.getElementById('counter').textContent = 'Counter: ' + counterValue;
    });

    // Display a greeting message
    const greetBtn = document.getElementById('greetBtn');
    greetBtn.addEventListener('click', function() {
      const userName = prompt('What is your name?');
      const messageDiv = document.getElementById('message');
      if (userName) {
        messageDiv.textContent = 'Hello, ' + userName + '! Welcome to the page.';
      } else {
        messageDiv.textContent = 'Hello, mysterious visitor!';
      }
    });
  </script>

```python
import React, { useState } from 'react';
import { Home, ArrowLeft } from 'lucide-react';

// Engineering fields data
const engineeringFields = [
  {
    id: 'mechanical',
    name: 'Mechanical Engineering',
    color: 'bg-blue-500',
    icon: '‚öôÔ∏è',
    facts: [
      {
        title: 'Mechanical Marvels',
        content: 'The world's largest mechanical engineering project is the International Space Station, a collaborative effort spanning multiple countries.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Robotics Revolution',
        content: 'Mechanical engineers are at the forefront of designing advanced robots for manufacturing, surgery, and space exploration.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Energy Efficiency',
        content: 'Modern mechanical engineering focuses on creating sustainable solutions like more efficient engines and renewable energy systems.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Fluid Dynamics',
        content: 'The study of fluid behavior powers innovations from aircraft design to blood flow analysis in medical devices.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Materials Science',
        content: 'Mechanical engineers develop new materials with specialized properties like self-healing polymers and ultra-strong composites.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Biomechanics',
        content: 'This field applies mechanical principles to biological systems, creating advanced prosthetics and understanding human movement.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Thermodynamics',
        content: 'The laws of energy transformation govern everything from power plants to refrigeration systems.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Nanotechnology',
        content: 'Mechanical engineering at microscopic scales enables medical devices smaller than a human cell.',
        image: '/api/placeholder/400/300'
      }
    ]
  },
  {
    id: 'electrical',
    name: 'Electrical Engineering',
    color: 'bg-yellow-500',
    icon: '‚ö°',
    facts: [
      {
        title: 'Quantum Computing',
        content: 'Electrical engineers are pioneering quantum computing, which could revolutionize data processing capabilities.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Smart Grids',
        content: 'Modern electrical engineering is creating intelligent power distribution networks that optimize energy consumption.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Renewable Energy',
        content: 'Solar and wind power technologies are pushing the boundaries of electrical engineering innovation.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Semiconductor Physics',
        content: 'The foundation of modern electronics, enabling transistors that power everything from smartphones to supercomputers.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Superconductivity',
        content: 'Materials with zero electrical resistance promise lossless power transmission and incredibly powerful electromagnets.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Signal Processing',
        content: 'Converting and filtering electrical signals enables modern wireless communication and audio/video technologies.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Electromagnetic Wave Theory',
        content: 'The basis for wireless communications, radar systems, and optical fiber networks that connect the world.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Power Electronics',
        content: 'Controlling and converting electrical power efficiently, essential for electric vehicles and renewable energy systems.',
        image: '/api/placeholder/400/300'
      }
    ]
  },
  {
    id: 'computer',
    name: 'Computer Engineering',
    color: 'bg-green-500',
    icon: 'üíª',
    facts: [
      {
        title: 'AI Breakthroughs',
        content: 'Computer engineers are developing increasingly sophisticated artificial intelligence systems that can learn and adapt.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Cybersecurity',
        content: 'Protecting digital infrastructure from cyber threats is a critical focus of modern computer engineering.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Internet of Things',
        content: 'Connecting everyday devices to the internet is transforming how we interact with technology.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Quantum Computing',
        content: 'Leveraging quantum mechanics to process information in ways that classical computers cannot, potentially breaking current encryption methods.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Distributed Systems',
        content: 'Networks of computers that work together as a unified system, powering cloud computing and global internet services.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Computer Architecture',
        content: 'The design of processor chips, memory systems and hardware accelerators that make computing devices faster and more efficient.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Edge Computing',
        content: 'Processing data closer to the source rather than in centralized data centers, reducing latency for critical applications.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Human-Computer Interaction',
        content: 'Designing intuitive interfaces between humans and computers, from touchscreens to brain-computer interfaces.',
        image: '/api/placeholder/400/300'
      }
    ]
  },
  {
    id: 'civil',
    name: 'Civil Engineering',
    color: 'bg-red-500',
    icon: 'üèóÔ∏è',
    facts: [
      {
        title: 'Structural Engineering',
        content: 'Creating buildings and bridges that can withstand natural disasters and support increasingly tall structures.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Transportation Systems',
        content: 'Designing efficient road networks, mass transit systems, and smart traffic management to connect communities.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Water Resources',
        content: 'Managing water systems from dams and reservoirs to treatment plants ensuring clean, accessible water.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Environmental Engineering',
        content: 'Developing sustainable practices and remediation techniques to protect natural ecosystems while supporting human needs.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Earthquake Engineering',
        content: 'Creating structures that can withstand seismic forces through specialized designs and materials.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Smart Cities',
        content: 'Integrating sensors and data analytics into urban infrastructure to improve efficiency and quality of life.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Geotechnical Engineering',
        content: 'Understanding soil and rock mechanics to create stable foundations for structures and prevent landslides.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Urban Planning',
        content: 'Designing cities that balance residential, commercial, and recreational spaces while promoting sustainability.',
        image: '/api/placeholder/400/300'
      }
    ]
  },
  {
    id: 'chemical',
    name: 'Chemical Engineering',
    color: 'bg-purple-500',
    icon: 'üß™',
    facts: [
      {
        title: 'Process Engineering',
        content: 'Designing and optimizing industrial processes to efficiently convert raw materials into valuable products.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Catalysis',
        content: 'Developing substances that accelerate chemical reactions without being consumed, critical for fuel production and pharmaceuticals.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Polymer Science',
        content: 'Creating new plastic materials with specialized properties for applications from medicine to aerospace.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Biochemical Engineering',
        content: 'Applying chemical engineering principles to biological systems for creating pharmaceuticals and biofuels.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Separation Processes',
        content: 'Developing methods to isolate specific components from mixtures, essential for purification in many industries.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Electrochemical Engineering',
        content: 'Designing batteries, fuel cells, and processes that convert between chemical and electrical energy.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Nanotechnology',
        content: 'Manipulating matter at atomic scales to create materials with unique properties for medicine, electronics, and energy.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Green Chemistry',
        content: 'Developing sustainable chemical processes that minimize waste and environmental impact while maximizing efficiency.',
        image: '/api/placeholder/400/300'
      }
    ]
  },
  {
    id: 'aerospace',
    name: 'Aerospace Engineering',
    color: 'bg-indigo-500',
    icon: 'üöÄ',
    facts: [
      {
        title: 'Propulsion Systems',
        content: 'Developing engines that can propel aircraft and spacecraft efficiently across Earth's atmosphere and into space.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Aerodynamics',
        content: 'Studying how air flows around objects to create more efficient aircraft designs and reduce fuel consumption.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Space Exploration',
        content: 'Creating vehicles and systems that can function in the vacuum of space and on other planetary bodies.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Satellite Technology',
        content: 'Designing spacecraft that orbit Earth to provide communications, navigation, weather monitoring, and scientific research.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Composite Materials',
        content: 'Developing lightweight yet strong materials that reduce aircraft weight while maintaining structural integrity.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Avionics',
        content: 'Creating the electronic systems that control aircraft and spacecraft navigation, communications, and operations.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Orbital Mechanics',
        content: 'Calculating the mathematics of spacecraft trajectories to navigate the solar system efficiently.',
        image: '/api/placeholder/400/300'
      },
      {
        title: 'Space Habitation',
        content: 'Designing systems that support human life in the extreme environment of space for long-duration missions.',
        image: '/api/placeholder/400/300'
      }
    ]
  }
];

// Main App Component
const EngineeringExplorer = () => {
  const [selectedField, setSelectedField] = useState(null);
  const [currentFact, setCurrentFact] = useState(0);

  // Home Screen Component
  const HomeScreen = () => (
    <div className="min-h-screen bg-gradient-to-br from-purple-100 to-blue-100 flex flex-col items-center justify-center p-4">
      <h1 className="text-4xl font-bold text-center mb-8 text-gray-800">
        Explore the World of Engineering
      </h1>
      <div className="grid grid-cols-1 sm:grid-cols-2 md:grid-cols-3 gap-6 w-full max-w-6xl">
        {engineeringFields.map((field) => (
          <div 
            key={field.id}
            onClick={() => setSelectedField(field)}
            className={`${field.color} text-white p-6 rounded-2xl shadow-lg transform transition hover:scale-105 cursor-pointer flex flex-col items-center`}
          >
            <span className="text-6xl mb-4">{field.icon}</span>
            <h2 className="text-2xl font-semibold text-center">{field.name}</h2>
            <p className="text-sm mt-2 text-white/80">{field.facts.length} concepts to explore</p>
          </div>
        ))}
      </div>
    </div>
  );

  // Field Details Screen Component
  const FieldDetailsScreen = () => {
    const handleScroll = (e) => {
      const { scrollTop, clientHeight, scrollHeight } = e.target;
      
      // Calculate current fact based on scroll position
      const newFact = Math.floor((scrollTop / (scrollHeight - clientHeight)) * (selectedField.facts.length - 1));
      if (newFact !== currentFact && newFact >= 0 && newFact < selectedField.facts.length) {
        setCurrentFact(newFact);
      }
    };

    return (
      <div className="min-h-screen relative">
        {/* Header with back button and title */}
        <div className="fixed top-0 left-0 right-0 bg-white/80 backdrop-blur-sm z-10 flex items-center p-4 shadow-sm">
          <button 
            onClick={() => setSelectedField(null)} 
            className="bg-gray-100 p-2 rounded-full hover:bg-gray-200 transition mr-4"
          >
            <ArrowLeft className="text-gray-800" size={20} />
          </button>
          <h1 className="font-bold text-xl text-gray-800">{selectedField.name}</h1>
          <div className="text-sm text-gray-500 ml-auto">
            {currentFact + 1}/{selectedField.facts.length}
          </div>
        </div>

        {/* Scrollable Content Container */}
        <div 
          onScroll={handleScroll}
          className="h-screen pt-16 overflow-y-scroll snap-y snap-mandatory"
        >
          {selectedField.facts.map((fact, index) => (
            <div 
              key={index} 
              className="h-screen flex flex-col items-center justify-center text-center p-6 snap-center"
              style={{
                backgroundColor: index % 2 === 0 ? '#f0f4f8' : '#e6eaf4'
              }}
            >
              <h2 className="text-3xl font-bold mb-6 text-gray-800">{fact.title}</h2>
              <img 
                src={fact.image} 
                alt={fact.title} 
                className="w-full max-w-lg mb-6 rounded-xl shadow-lg"
              />
              <p className="text-xl max-w-2xl text-gray-700">{fact.content}</p>
            </div>
          ))}
        </div>
      </div>
    );
  };

  return (
    <div className="font-sans">
      {selectedField ? <FieldDetailsScreen /> : <HomeScreen />}
    </div>
  );
};

export default EngineeringExplorer;
```

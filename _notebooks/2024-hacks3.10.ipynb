```python
#Long python hack
# function to show the current contents of the list
def show_list(lst):
    print("Here’s the current list:", lst)

# function to insert an item into the list
def insert_item(lst, item):
    lst.append(item)  # add the item at the end of the list
    print(f"'{item}' was added.")

# function to delete an item from the list
def delete_item(lst, item):
    if item in lst:
        lst.remove(item)  # remove the item if it exists
        print(f"'{item}' was removed.")
    else:
        print(f"'{item}' wasn’t found in the list.")

# function to check if a given item is in the list
def verify_item(lst, item):
    if item in lst:
        print(f"'{item}' is indeed in the list.")
    else:
        print(f"'{item}' is not in the list.")

# function to identify the largest number in the list
def get_max(lst):
    highest = lst[0]  # assume the first element is the largest
    for num in lst:
        if num > highest:
            highest = num
    return highest

# function to identify the smallest number in the list
def get_min(lst):
    lowest = lst[0]  # assume the first element is the smallest
    for num in lst:
        if num < lowest:
            lowest = num
    return lowest

# initializing an empty list
current_list = []

# predefined actions (simulating input actions)
tasks = [
    ('insert', 'apple'),
    ('insert', 'banana'),
    ('verify', 'banana'),
    ('delete', 'apple'),
    ('show', None),
    ('insert', 10),
    ('insert', 5),
    ('show', None),
    ('get_max', None),
    ('get_min', None),
    ('exit', None)
]

# performing the operations as per the tasks list
for task, val in tasks:
    if task == 'insert':
        insert_item(current_list, val)
    elif task == 'delete':
        delete_item(current_list, val)
    elif task == 'verify':
        verify_item(current_list, val)
    elif task == 'show':
        show_list(current_list)
    elif task == 'get_max':
        if current_list and all(isinstance(x, (int, float)) for x in current_list):
            print("The max number is:", get_max(current_list))
        else:
            print("list is empty or non-numeric.")
    elif task == 'get_min':
        if current_list and all(isinstance(x, (int, float)) for x in current_list):
            print("the min number is:", get_min(current_list))
        else:
            print("list is empty or non-numeric.")
    elif task == 'exit':
        print("shutting down this program.")
        break
```

```python
// initialize an empty shopping cart array
let cart = [];

// function to add an item to the cart
function addItem(item) {
    cart.push(item);  // add the item to the cart
    console.log(`${item} has been added to your cart.`);
    displayCart();  // show the updated cart
}

// function to remove an item from the cart
function removeItem(item) {
    let index = cart.indexOf(item);  // find the index of the item
    if (index > -1) {
        cart.splice(index, 1);  // remove the item if it exists
        console.log(`${item} has been removed from your cart.`);
    } else {
        console.log(`${item} is not in your cart.`);
    }
    displayCart();  // show the updated cart
}

// function to display the cart items
function displayCart() {
    if (cart.length === 0) {
        console.log("Your cart is empty.");
    } else {
        console.log("Your cart contains:", cart);
    }
}

// simulate adding and removing items from the cart
addItem('apple');  // add 'apple' to the cart
addItem('banana');  // add 'banana' to the cart
removeItem('apple');  // remove 'apple' from the cart
addItem('orange');  // add 'orange' to the cart
displayCart();  // display the final state of the cart
```

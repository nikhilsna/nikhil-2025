```python
import random

#Popcorn hack 1

def magic_8_ball():
    random_num = random.random()
    
    if random_num < 0.5:
        return "Y"
    elif random_num < 0.75:
        return "N"
    else:
        return("try later man.")

# final test
for i in range(10):
    print(f"8-ball says: {magic_8_ball()}")

#Actual Hw

import random

def roll_dice():
    """Roll two dice and return their values and sum."""
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    total = die1 + die2
    return die1, die2, total

def play_dice_game():
    """
    Play one round of the dice game.
    Returns True if player wins, False if player loses.
    """
    # First roll
    die1, die2, total = roll_dice()
    print(f"You rolled {die1} + {die2} = {total}")
    
    # Check for win or loss on first roll
    if total in [7, 11]:
        print("You win!")
        return True
    elif total in [2, 3, 12]:
        print("You lose!")
        return False
    
    # If no immediate win/loss, establish the point
    point = total
    print(f"Your point is {point}. Continue rolling until you get {point} again (win) or 7 (lose).")
    
    # Continue rolling until win or loss
    while True:
        input("Press Enter to roll again...")
        die1, die2, total = roll_dice()
        print(f"You rolled {die1} + {die2} = {total}")
        
        if total == point:
            print(f"You rolled your point ({point})! You win!")
            return True
        elif total == 7:
            print("You rolled a 7. You lose!")
            return False

def main():
    """Main game function with game loop and statistics."""
    wins = 0
    losses = 0
    
    while True:
        choice = input("\nDo you want to play? (yes/no): ").lower()
        
        if choice in ['yes', 'y']:
            result = play_dice_game()
            if result:
                wins += 1
            else:
                losses += 1
            print(f"Statistics - Wins: {wins}, Losses: {losses}")
        elif choice in ['no', 'n']:
            break
        else:
            print("Invalid input. Please enter 'yes' or 'no'.")
    
    # Show final statistics
    total_games = wins + losses
    if total_games > 0:
        win_percentage = (wins / total_games) * 100
    else:
        win_percentage = 0
        
    print("\nFinal Statistics:")
    print(f"Games Played: {total_games}")
    print(f"Wins: {wins} ({win_percentage:.1f}%)")
    print(f"Losses: {losses} ({100 - win_percentage:.1f}%)")
    print("Thanks for playing!")

if __name__ == "__main__":
    print("Welcome to the Dice Game!")
    print("Rules:")
    print("1. Roll two dice")
    print("2. If you roll 7 or 11 on your first roll, you win")
    print("3. If you roll 2, 3, or 12 on your first roll, you lose")
    print("4. If you roll any other number, that becomes your 'point'")
    print("5. Keep rolling until you either roll your point again (win) or roll a 7 (lose)")
    main()
```
